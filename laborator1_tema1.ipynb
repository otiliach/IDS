1. (1p) (fizz-buzz test) Sa se scrie numerele de la 1 la n; pentru fiecare multiplu de 3 se va scrie in locul numarului 'Fizz', pentru multiplu al lui 5 se va scrie 'Buzz'; daca numarul este multiplu de 15 se va scrie in locul lui 'FizzBuzz'.


````python
n = input("Enter a number: ")
for i in range(1, int(n)):
    if i%15==0:
        print("FizzBuzz")
    elif i%3==0:
        print("Fizz")
    elif i%5==0:
        print("Buzz")
    else: print(i)
````

2. (1p) Sa se verifice [conjectura Collatz](https://en.wikipedia.org/wiki/Collatz_conjecture) pentru numerele intre 1 si 1000. 

````python
for n in range(2,1000):
    print(n,"->",end=' ')
    while n!=1:
        if n%2==0:
            n=n/2
        else: 
            n=n*3+1
    print(n)
````

6. (2p) Se da un dictionar in care cheile sunt nume de studenti, iar valorile sunt liste de carti preferate. Sa se determine:
    1. Care sunt studentii care au in lista de preferinte o carte specificata?
    2. Care sunt perechile de studenti care au aceleasi preferinte de carti (ordinea in lista de preferinte nu e relevanta)
    3. Pentru un student dat, care sunt studentii cu care are cele mai multe carti comune in liste de preferinte; daca sunt mai multi astfel de studenti cei mai apropiati, se vor enumera toti.
    4. Pentru doi studenti, A si B, care e lista de stergeri si adaugari prin care lista lui A devine identica cu lista lui B (ordinea in lista nefiind importanta)
    5. Toate perechile de studenti A, B pentru care lista de preferinte a lui A este inclusa in (dar nu coincide cu) lista de preferinte a lui B. 
    
Indicatie: considerati tipul de date `set`.

Exemplu:
```python
preferinte = {
    'Popescu': ['carte2', 'carte1', 'carte3', 'carte4'],
    'Ionescu': ['carte1', 'carte2', 'carte3', 'carte7', 'alta carte'],
    'Georgescu': ['carte1', 'alta carte'],
    'Xulescu': ['carte2', 'carte5', 'carte6', 'carte7'],
    'Dragomir': ['carte4', 'carte1', 'carte2', 'carte3']
}
```
    1. carte = 'alta carte' -> ['Ionescu', 'Georgescu']
    2. Popescu si Dragomir au aceleasi preferinte: ['carte2', 'carte1', 'carte3', 'carte4']. Se remarca faptul ca nu se afiseaza si perechea echivalenta: Dragomir si Popescu.
    3. Xulescu -> Suprapunere maxima de preferinte: 2 cu colegii: ['Ionescu']
    4. A='Popescu', B='Ionescu' -> Se vor adauga cartile: {'carte7', 'alta carte'}, se vor scoate cartile: {'carte4'}
    5. Preferintele lui Georgescu sunt subset propriu al preferintelor lui Ionescu

````python
preferinte = {
    'Popescu': ['carte2', 'carte1', 'carte3', 'carte4'],
    'Ionescu': ['carte1', 'carte2', 'carte3', 'carte7', 'alta carte'],
    'Georgescu': ['carte1', 'alta carte'],
    'Xulescu': ['carte2', 'carte5', 'carte6', 'carte7'],
    'Dragomir': ['carte4', 'carte1', 'carte2', 'carte3']
}
#Care sunt studentii care au in lista de preferinte o carte specificata?
carte='alta carte'
studenti=set()
for student, preferinte_student in preferinte.items():
    if carte in preferinte_student:
        studenti.add(student)
print(studenti)
#Care sunt perechile de studenti care au aceleasi preferinte de carti (ordinea in lista de preferinte nu e relevanta)
studenti = list(preferinte.keys())
studenti_cu_preferinte_comune=set()
for student1 in preferinte:
    for student2 in preferinte:
        if student1 != student2:
            if len(preferinte[student1])==len(preferinte[student2]):
                preferinte_comune=set(preferinte[student1]).intersection(set(preferinte[student2]))
                if len(preferinte_comune)==len(preferinte[student1]):
                    if (student1, student2) not in studenti_cu_preferinte_comune and (student2, student1) not in studenti_cu_preferinte_comune:
                        print(student1,' si ',student2,' au aceleasi preferinte:',preferinte_comune)
                        studenti_cu_preferinte_comune.add((student1, student2))
       
        
#Pentru un student dat, care sunt studentii cu care are cele mai multe carti comune in liste de preferinte; daca sunt mai multi astfel de studenti cei mai apropiati, se vor enumera toti.
student_dat = 'Xulescu'
preferinte_student_dat = set(preferinte[student_dat])
nr_carti_comune = 0
studenti_cu_carti_comune = set()

for student, preferinte_student in preferinte.items():
    if student != student_dat:
        carti_comune = len(preferinte_student_dat.intersection(set(preferinte_student)))
        if carti_comune > nr_carti_comune:
            nr_carti_comune = carti_comune
            studenti_cu_carti_comune = {student}
        elif carti_comune == nr_carti_comune:
            studenti_cu_carti_comune.add(student)

print('Studentul cu cele mai multe carti comune cu ',student_dat,end=': ')
for student in studenti_cu_carti_comune:
    print(student)

#Pentru doi studenti, A si B, care e lista de stergeri si adaugari prin care lista lui A devine identica cu lista lui B (ordinea in lista nefiind importanta)
A = 'Popescu' 
B = 'Ionescu'
lista_stergeri=set(preferinte[A]).difference(set(preferinte[B]))
lista_adaugari=set(preferinte[B]).difference(set(preferinte[A]))
print(f"{A}, {B} -> Se vor adauga cartile: {lista_adaugari}, se vor scoate cartile:  {lista_stergeri}")
 #A='Popescu', B='Ionescu' -> Se vor adauga cartile: {'carte7', 'alta carte'}, se vor scoate cartile: {'carte4'}
#Toate perechile de studenti A, B pentru care lista de preferinte a lui A este inclusa in (dar nu coincide cu) lista de preferinte a lui B
for student_a in preferinte:
    for student_b in preferinte:
        if student_a != student_b:
            if len(preferinte[student_a])!=len(preferinte[student_b]):
                if len(preferinte[student_a])==len(set(preferinte[student_a]).intersection(set(preferinte[student_b]))):
                    print(f"Preferintele lui {student_a} sunt subset propriu al preferintelor lui {student_b}")
````

7. (1 p) Se pleaca de la un numar intreg. Fiecare cifra a sa se scrie cu litere, in limba engleza (de exemplu 5->five). Calculati numarul total de caractere rezultate, iar pentru numarul obtinut repetati procedura. Verificati pentru numerele de la 1 la n ca se obtine intr-un numar finit de pasi numarul 4. Exemplu: 123->onetwothree->11->oneone->6->six->3->three->5->five->4->four->4->four....

````python
cifre= ['', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
n = int(input("Enter a number: "))
for i in range(1,n+1):
    while i != 4:
        numar_cuvant = ''
        for cifra in str(i):
            numar_cuvant += cifre[int(cifra)]
        print(f"{i} -> {numar_cuvant}->",end=' ')
        i = len(numar_cuvant)
    print(f"{i} -> {cifre[i]}")
    
````
