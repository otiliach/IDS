```python
import panel as pn
import hvplot.pandas
import pandas as pd
import numpy as np
import csv
pn.extension('tabulator')
```

```python
pn.extension(design='material')
csv_file = ("census_income.csv")

data:pd.DataFrame = pd.read_csv(csv_file)
```

```python
data
```

```python
data.columns
```

<h1>0. Preprocesare</h1>

```python
import re


def convert_string_to_int(item:str):
    '''
    Converteste un tip de data String la int
    :param item(string): un sir introdus ca parametru
    :return match(int):returnează sirul potrivit de regex ca valoare convertita in int
    '''
    match = re.search(r'\d+', item)
    return int(match.group())

print(data.dtypes) 
data:pd.DataFrame = data.replace(',','', regex=True)
data['age']:pd.Series = data['age'].apply(convert_string_to_int)
data['education-num']:pd.Series = data['education-num'].apply(convert_string_to_int)   
data['hours-per-week']:pd.Series = data['hours-per-week'].apply(convert_string_to_int)
data['capital-gain']:pd.Series = data['capital-gain'].apply(convert_string_to_int)
data['capital-loss']:pd.Series = data['capital-loss'].apply(convert_string_to_int)    
```

```python
print(data.dtypes)
```

```python
data
```

```python
idf=data.interactive()
```

<h1>1.Filter by Age</h1>

```python
min_age:int=data.age.min()
max_age:int=data.age.max()
age_slider:pn.widgets.IntSlider=pn.widgets.IntSlider(name="Age",start=min_age,end=max_age,step=1, value=30)

range_slider:pn.widgets.RangeSlider = pn.widgets.RangeSlider(
    name='Age Slider', start=min_age, end=max_age, value=(20, 80), step=1)

#range_slider
```

```python
age_slider
```

```python
yaxis_df:pn.widgets.RadioButtonGroup =pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['hours-per-week', 'capital-gain'],
    button_type='success'
)
```

```python
educations:list=['Bachelors', 'Some-college', '11th', 'HS-grad', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', '9th', '7th-8th', '12th', 'Masters', '1st-4th+', '10th', 'Doctorate', '5th-6th', 'Preschool']
income_pipeline:hvplot.interactive.Interactive = (
    idf[
        (idf.age >= age_slider) &
        (idf.education.isin(educations))
    ]
    .groupby(['education', 'age'])[yaxis_df].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='age')
    .reset_index(drop=True)
)
income_pipeline
```

<h1>1. Grafice în funcție de vârstă<h1>

```python
plot1:hvplot.interactive.Interactive=income_pipeline.hvplot.line(x='age',by='education', y=yaxis_df,title="income pipeline")

plot1
```

<h1>2. Table - capital gain </h1>

```python
income_table:hvplot.interactive.Interactive=income_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size=10, sizing_mode='stretch_width')
income_table
```

<h1>3. Capital-gain vs capital-loss scatterplot </h1>

```python
gain_vs_loss_scatterplot_pipeline:hvplot.interactive.Interactive = (
    idf[
        (idf.age== age_slider) &
        (~(idf.education.isin(educations)))
    ]
    .groupby(['education', 'age', 'capital-loss'])['capital-gain'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='age')
    .reset_index(drop=True)
)
```

```python
gain_vs_loss_scatterplot_pipeline
```

<h1>3.1 Bar chart: Capital by marital status </h1>

```python
yaxis_capital:pn.widgets.select.RadioButtonGroup=pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['capital-loss','capital-gain','hours-per-week'],
    button_type='success'
)

marital_status_data:list=['Married-civ-spouse', 'Divorced', 'Never-married', 'Separated', 'Widowed', 'Married-spouse-absent', 'Married-AF-spouse']

source_bar_pipeline:hvplot.interactive.Interactive = (
    idf[
        (idf.age == age_slider) &
        (idf['marital-status'].isin(marital_status_data))
    ]
    .groupby(['age', 'marital-status'])[yaxis_capital].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='age')
    .reset_index(drop=True)
)
```

```python
source_bar_plot: hvplot.interactive.Interactive = source_bar_pipeline.hvplot(kind='bar',x='marital-status', y=yaxis_capital,title="Data by marital status ")
source_bar_plot
```
